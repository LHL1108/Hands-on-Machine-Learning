# 手动实现K-NN 

## K-NN 与 K-Means

## K-NN伪代码 

对待预测数据中的每个点依次执行以下操作:

    1.计算训练集中的每个点与当前点之间的距离
    
    2.按照距离递增次序排序
    
    3.选取当前点距离最小的k个点
    
    4.确定前k个点所在类别的出现频率
    
    5.返回出现频率最高的类别作为当前新数据的预测分类
    

    

## K-NN 的手动实现

```python
from numpy import *
import operator

group = array([[1.0, 1.1], [1.0, 1.0], [0, 0], [0, 0.1]])
labels = ['A', 'A', 'B', 'B']
inX = [0,0]
k = 3

dataSetSize = group.shape[0]
a = (tile(inX,(dataSetSize, 1)) - group)
b = a**2
c = b.sum(axis = 1)
d = c**0.5
d
# array([1.48660687, 1.41421356, 0.        , 0.1       ])

#按值从小到大的顺序返回index
d.argsort()
#array([2, 3, 1, 0], dtype=int64)

classCount = {} 
for i in range (k):
    #依次返回前k个数据的label（B、B、A）
    klabel = labels[d.argsort()[i]]
    classCount[klabel] = classCount.get(klabel, 0) + 1
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)
    #[('B', 1)]   [('B', 2)]    [('B', 2), ('A', 1)]
    
    #返回出现次数最多的标签(sortedClassCount的第0行第0列)
sortedClassCount[0][0]
```
